

#Using apply function to create a new Column called (Healthy). The function should check the calories values in Calories column and return the following categorical values:

#a) 0 < calories <= 300 ====> return 1 

#b) 301 < calories <= 600====> return 2

#c) 601 < calories <= 900====> return 3

#d) calories > 901 ====> return 4


```python
def Get_Calories (row):
    
    calories = row["Calories"]
   
    if 0 < calories <= 300:
        
        return 1
    elif 301 < calories <= 600:
 
        return 2
    elif 601 < calories <= 900:
        
        return 3
    elif calories > 901:
        
        return 4
```

```python
df["Healthy"] = df.apply(Get_Calories, axis=1)
```

#In the Serving Size column, the size of the meals is shown in oz and gram. Use apply function and create a new Column that return the value size of meal using Gram unit only..

```python
def Get_Titles(name):
    parse_name = name.split(" ")
    for i in range(len(parse_name)):
        if "," in parse_name[i]:
            return parse_name[1 + i]
    return None
```

```python
df3 = df["Serving Size"].apply(Get_Titles)
```

```python
df["Gram_Unit"] = df3
df = df.drop("Gram_Unit", axis=1)
```
