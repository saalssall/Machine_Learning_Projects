```python
import pandas as pd
import numpy as np
```

```python
# Reading the three excel sheets using pandas
orders = pd.read_excel("carWorkshop-1.xlsx", sheet_name=0)
lic_plates = pd.read_excel("carWorkshop-1.xlsx", sheet_name=1)
services = pd.read_excel("carWorkshop-1.xlsx", sheet_name=2)
```

```python
# Changing the 'Requested On' column type to date
orders["Requested On"] = pd.to_datetime(orders["Requested On"])
```

```python
# Renaming plate column to join it with the plate sheet
orders.rename({"LIC PLATE": "LICENCE PLATE"}, axis=1, inplace=True)
```

```python
# Correcting the spelling mistakes of the orders sheet
orders.replace({"TIRiS": "TIRES", "OIL/FALTER": "OIL/FILTER"}, inplace=True)
```

```python
# Removing the 'lic' prefix to use it as an ID and merge with the order page
lic_plates["LICENCE PLATE"] = lic_plates["LICENCE PLATE"].str[4:]
```

```python
# Combining the 3 sheets using the 'lic_plate' and 'service' columns
combined = orders.merge(lic_plates, on="LICENCE PLATE").merge(services, on="SERVICE", how="outer")
```

```python
# Dropping duplicates and rows with no assigned mechanics
combined.drop_duplicates(inplace=True)
combined.dropna(subset=["MECH"], inplace=True)
```

```python
# Resetting the index after deleting problematic rows
combined.reset_index(drop=True, inplace=True)
combined
```
