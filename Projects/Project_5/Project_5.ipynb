#Using apply function, customize your own algorithm to discretize the age of passengers into (Child, Young and Old), and assign it to a new column? Hint: Choose the range of age yourself

```python
#Elderly 
def Get_Old_Pass(row):
    
    Age = row["Age"]
    
    if (Age>60):
        
        return "Old"
    elif(18 <= Age <= 60):
        
        return "Young"
    elif(Age<18):
        
        return "Child"
```

```python
df["Elderly"] = df.apply(Get_Old_Pass, axis=1)
```
